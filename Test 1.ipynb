```python
1. #Does string 's' contain string 't' ?
def contains(s, t):
    # Base cases
    if len(t) == 0:
        return True
    
    if len(s) == 0:
        return False
    
    # If the first character of s matches the first character of t
    if s[0] == t[0]:
        # Recursively check if the remaining characters of s and t also match in order
        return contains(s[1:], t[1:])
    
    # If the characters don't match, move to the next character in s and keep t unchanged
    return contains(s[1:], t)

s = input()
t = input()

ans = contains(s, t)
if ans is True:
    print('true')
else:
    print('false')
  
```

```python
#2. Maximum Profit on App
def maximumProfit(arr):

    arr.sort()  # Sort the budgets in ascending order
    N = len(arr)
    max_profit = 0
    for i in range(N):
        curr_profit = arr[i] * (N - i)  # Calculate profit for each price
        if curr_profit > max_profit:
            max_profit = curr_profit
    return max_profit

n = int(input())
arr = [int(ele) for ele in input().split()]
ans = maximumProfit(arr)
print(ans)
```
